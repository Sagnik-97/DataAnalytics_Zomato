```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file with ISO-8859-1 encoding
df = pd.read_csv('zomato.csv', encoding='ISO-8859-1')



# Calculate the number of restaurants in each price range
num_restaurants_by_price_range = df.groupby(
    'Price range')['Restaurant ID'].count()

# Plot a vertical bar graph of number of restaurants by price range
plt.figure(figsize=(8, 6))
plt.bar(num_restaurants_by_price_range.index,
        num_restaurants_by_price_range.values, color='purple')
plt.title('Number of Restaurants by Price Range', fontsize=16)
plt.xlabel('Price Range', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()





```

```python
# Calculate the number of restaurants in each locality
num_restaurants_by_locality = df.groupby(
    'Locality')['Restaurant ID'].count().sort_values(ascending=False).head(10)

# Plot a vertical bar graph of number of restaurants by locality
plt.figure(figsize=(8, 6))
plt.bar(num_restaurants_by_locality.index,
        num_restaurants_by_locality.values, color='green')
plt.title('Number of Restaurants by Locality (Top 10)', fontsize=16)
plt.xlabel('Locality', fontsize=12)
plt.ylabel('Number of Restaurants', fontsize=12)
plt.xticks(rotation=90, fontsize=12)
plt.yticks(fontsize=12)
plt.show()

```

```python
# Calculate the number of restaurants that have table booking
num_restaurants_with_table_booking = df['Has Table booking'].value_counts()

# Plot a pie chart of number of restaurants with table booking
plt.figure(figsize=(8, 6))
plt.pie(num_restaurants_with_table_booking.values,
        labels=num_restaurants_with_table_booking.index, autopct='%1.1f%%', colors=['red', 'orange'])
plt.title('Percentage of Restaurants with Table Booking')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python

# Filter the data to include only restaurants in India
df_india = df[df['Country Code'] == 1]

# Top 10 cities with the most restaurants
top_cities = df_india['City'].value_counts().nlargest(10)

# Bar graph for top 10 cities with the most restaurants
plt.figure(figsize=(8, 6))
plt.bar(top_cities.index, top_cities.values, color='orange')
plt.title('Top 10 Cities with the Most Restaurants')
plt.xlabel('City')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()


```

```python
# Average rating by city
avg_rating_by_city = df_india.groupby(
    'City')['Aggregate rating'].mean().sort_values(ascending=False)

# Bar graph for average rating by city
plt.figure(figsize=(8, 6))
plt.bar(avg_rating_by_city.index, avg_rating_by_city.values, color='blue')
plt.title('Average Rating by City')
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Number of restaurants with online delivery
delivery = df_india['Has Online delivery'].value_counts()

# Pie chart for number of restaurants with online delivery
plt.figure(figsize=(8, 6))
plt.pie(delivery.values, labels=delivery.index,
        autopct='%1.1f%%', startangle=90)
plt.title('Number of Restaurants with Online Delivery')
plt.tight_layout()
plt.show()
```

```python
# Top 10 cuisines with the most number of restaurants
top_cuisines = df_india['Cuisines'].str.split(
    ', ', expand=True).stack().value_counts().nlargest(10)

# Bar graph for top 10 cuisines with the most number of restaurants
plt.figure(figsize=(8, 6))
plt.bar(top_cuisines.index, top_cuisines.values, color='green')
plt.title('Top 10 Cuisines with the Most Number of Restaurants')
plt.xlabel('Cuisine')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
